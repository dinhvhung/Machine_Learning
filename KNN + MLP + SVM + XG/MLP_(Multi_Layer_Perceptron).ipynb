```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import warnings
warnings.filterwarnings('ignore')

print("🚀 Khởi động mô hình Mạng Nơ-ron (MLP Regressor)...")

# 1. Chuẩn bị dữ liệu
X = df.drop('median_house_value', axis=1)
y = df['median_house_value']

# 2. Khởi tạo và Huấn luyện mô hình qua 5 tỷ lệ chia dữ liệu
print("🔥 Đang kiểm thử mô hình Neural Network qua 5 tỷ lệ chia Train:Test...")

ratios = {'5:5': 0.5, '6:4': 0.4, '7:3': 0.3, '8:2': 0.2, '9:1': 0.1}
results = []

# Cố định cấu hình MLP tối ưu
# Sử dụng 3 lớp ẩn, hàm kích hoạt relu và cơ chế early_stopping chống học vẹt
nn_model = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),
    activation='relu',
    solver='adam',
    max_iter=300,
    early_stopping=True,
    random_state=42
)

for label, test_size in ratios.items():
    print(f"🔄 Đang quét tỷ lệ {label}...")

    # Tách tập Train/Test theo tỷ lệ
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Chuẩn hóa Z-score
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Nén giá nhà bằng Logarit
    y_train_log = np.log1p(y_train)
    y_test_log = np.log1p(y_test)

    # Huấn luyện
    nn_model.fit(X_train_scaled, y_train_log)

    # Dự đoán
    y_pred_log = nn_model.predict(X_test_scaled)

    y_pred_log = np.clip(y_pred_log, a_min=0, a_max=16)

    # Bung Logarit
    y_pred_real = np.expm1(y_pred_log)
    y_test_real = np.expm1(y_test_log)

    # Tính toán các chỉ số đánh giá
    r2 = r2_score(y_test_real, y_pred_real)
    rmse = np.sqrt(mean_squared_error(y_test_real, y_pred_real))
    mae = mean_absolute_error(y_test_real, y_pred_real)
    mape = np.mean(np.abs((y_test_real - y_pred_real) / y_test_real)) * 100

    results.append({
        'Tỷ lệ Train:Test': label,
        'R² Score': round(r2, 4),
        'RMSE ($)': round(rmse, 0),
        'MAE ($)': round(mae, 0),
        'MAPE (%)': round(mape, 2)
    })

# 3. Trình bày kết quả
results_df = pd.DataFrame(results)

print("\n" + "="*60)
print("🏆 BẢNG KẾT QUẢ HIỆU NĂNG NEURAL NETWORK QUA 5 TỶ LỆ DỮ LIỆU")
print("="*60)
display(results_df)
```
